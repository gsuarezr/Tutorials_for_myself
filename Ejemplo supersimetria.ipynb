```python
from sympy import *
```

```python
k1,k2,x,z,alpha=symbols('k1 k2 x z alpha',real=True)
```

```python
#u=cosh(k1*x)*exp(I * k1**2 * z ) +alpha*cosh(k2*x)*exp(I * k2**2 *z)
u=cosh(k1*x)*exp(I * k1**2 * z ) +I*alpha*sinh(k2*x)*exp(I * k2**2 *z)
l1=0
u
```

```python
def VL(Vo,l1,u):
    V=Vo+I*diff(log(l1),z) -2 * diff(log(u),x,2)
    V=simplify(V)
    return V
```

```python
V=VL(0,0,u)
reV=re(V)
imV=im(V)
```

```python
from sympy.plotting import plot3d
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
plot3d(reV.subs({k1:0.4, k2: 0.1,alpha:0.5}),(x,-15,15),(z,-60,60),xlabel=r'$\theta_{1}$'
               ,ylabel=r'$\theta_{2}$',title=r'$Probablity$')
```

```python
plot3d(imV.subs({k1:0.4, k2: 0.1,alpha:0.5}),(x,-15,15),(z,-60,60),xlabel=r'$\theta_{1}$'
               ,ylabel=r'$\theta_{2}$',title=r'$Probablity$')
```

```python
#v= sinh(k1*x)* exp(I * k1**2 * z) + alpha*sinh(k2*x)*exp(I* k2 **2 *z)
v=sinh(k1*x)* exp(I * k1**2 * z) + I*alpha*cosh(k2*x)*exp(I* k2 **2 *z)
v
```

```python
def L(u,v):
    Vn=diff(v,x)- diff(u,x) * (v/u)
    return Vn

Vn=L(u,v)
simplify(Vn)
```

```python
Vn= Vn*conjugate(Vn)
```

```python
plot3d(Vn.subs({k1:0.4, k2: 0.1,alpha:0.5}),(x,-15,15),(z,-60,60),xlabel=r'$\theta_{1}$'
               ,ylabel=r'$\theta_{2}$',title=r'$Probablity$')
```
