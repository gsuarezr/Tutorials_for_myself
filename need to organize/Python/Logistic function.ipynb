```python
def logistic(r,x0,n):
    x=0
    for i in range(0,n):
        x=r*x0*(1-x0)
        x0=x
    return x
```

```python
logistic(2.6,0.2,10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=np.linspace(1,10000,10000)
a=np.array([])
for i in range(1,10001):
    a=np.append(a,logistic(3.72,0.2,i)-logistic(3.72,0.200001,i))
```

```python
a=-1*a
```

```python
plt.scatter(n,a)
plt.ylim(0,1)
plt.show()
```
