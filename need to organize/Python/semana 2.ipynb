```python
%matplotlib notebook
import math 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
promedio=0
std_desviacion=1
N=100000
np.random.seed(0)
X=promedio+std_desviacion*np.random.randn(N)
plt.ylim(-10,10)
plt.xlabel(r'$n$',fontsize=18)
plt.ylabel(r'$X_n$',fontsize=18)
plt.plot(X,',')
plt.show()
```

```python
plt.hist(X,bins=25,normed=True)
plt.show()
```

```python
x=np.arange(-10,10,0.01)
y=np.exp(-(x-promedio)**2/(2*std_desviacion**2))/ np.sqrt(2* np.pi* std_desviacion**2)
plt.hist(X,bins=25,normed=True)
plt.plot(x,y)
plt.show()
```

```python
def autocorrelacion(n):
    correlacion=np.correlate(n,n,mode='full')
    return correlacion[N-1:]
```

```python
c=np.zeros(N)
c=autocorrelacion(X- promedio)/N
plt.plot(c)
plt.show()
```

# tarea realizar el mismo analisis para una distribucion binomial y de poisson


```python
ave = 2.0
std = np.sqrt(4.0)
N = 1000
np.random.seed(1000)
X = np.random.normal(ave,std,N)
x=np.arange(-10,10,0.01)
y=np.exp(-(x-ave)**2/(2*std**2))/ np.sqrt(2* np.pi* std**2)
plt.hist(X,bins=25,normed=True)
plt.plot(x,y)
plt.show()
```

```python
p=0.5
M=1000
N=100000
ave=p*M
std=np.sqrt(M*p*(1-p))
print('p=',p,'M=',M)
np.random.seed(0)
X=np.random.binomial(M,p,N)
nmin=np.int(ave-std*5)
nmax=np.int(ave+std*5)
nbin=nmax-nmin+1
plt.hist(X,range=[nmin,nmax],bins=nbin,normed=True)
x=np.arange(nmin,nmax,0.01/std)
y=np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2)
plt.plot(x,y,color='b')
plt.show()
```

```python
M=4
N=100000
ave=M/2
std=np.sqrt(M/12)
print('M=',M)
np.random.seed(0)
X=np.zeros(N)
for i in range(N):
    X[i] += np.sum(np.random.rand(M))
nmin=np.int(ave-std*5)
nmax=np.int(ave+std*5)
plt.hist(X,range=[nmin,nmax],bins=50,normed=True)
x=np.arange(nmin,nmax,0.01/std)
y=np.exp(-(x-ave)**2/(2*std**2))/np.sqrt(2*np.pi*std**2)
plt.plot(x,y,color='b')
plt.show()
```

```python
def rng(seed, a, b, M, ntotal):
    data = np.zeros(ntotal)
    data[0] = seed
    for i in range(1,ntotal):
        data[i] = np.mod((a*data[i-1]+b), M)
    return data/np.float(M)
```

```python
N=5*10**5
promedio=0
X=rng(123456,8121,28411,134456,N)
for i in range(0,N):
    promedio+=X[i]
promedio=promedio/N
```

```python
def autocorrelacion(n):
    correlacion=np.correlate(n,n,mode='full')
    return correlacion[N-1:]
```

```python
c=np.zeros(8000)
X=X[0:8000]
c=autocorrelacion(X- promedio)/8000
plt.plot(c)
plt.show()
```
