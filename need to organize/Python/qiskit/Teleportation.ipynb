```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
circuit= QuantumCircuit(3,3)
circuit.draw(output='mpl')
```

```python
circuit.x(0)# pasamos el primer qubit de 0 a 1
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)# aplicamos una hadamad y una Cnot para obtener entrelazamiento
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()

circuit.cx(1,2)
circuit.cz(0,2)
circuit.measure(2,2)

circuit.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator,shots=1024).result()
counts=result.get_counts()
plot_histogram(counts)
```
