```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r,x0):
   return r*x0*(1-x0)
```

```python
x = np.linspace(0, 1)
plt.plot(x,logistic(2,x))
```

```python
def cobwebb(r, x0, n, ax=None):
    # graficamos la linea y la parabola
    # y=x linea diagonal.
    t = np.linspace(0, 1)
    plt.plot(t, logistic(r, t), 'k', lw=2)
    plt.plot([0, 1], [0, 1], 'k', lw=2)

    #iteradamente aplicamos y=f(x) y graficamos las siguientes lineas:
    # (x, x) -> (x, y)
    # (x, y) -> (y, y)
    x = x0
    for i in range(n):
        y = logistic(r, x)
        # graficamos ambas lineas.
        plt.plot([x, x], [x, y], 'k', lw=1)
        plt.plot([x, y], [y, y], 'k', lw=1)
        # Graficamos las posiciones con mayor opacidad cada vez solo por estetica
        plt.plot([x], [y], 'ok', ms=10,alpha=(i + 1) / n)
        x = y

    plt.title(f"$r={r:.1f}, \, x_0={x0:.1f}$")


cobwebb(3.5, .1, 1000, ax=ax1)
```
