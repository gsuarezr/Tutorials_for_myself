```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sympy import *
```

```python
init_printing(use_unicode=True)
N, n, t = symbols('N  n t ')
```

```python
alpha =pi/(2*N)
```

```python
B=Matrix([[cos(alpha),sin(alpha)],[-sin(alpha),cos(alpha)]])
B
```

```python
#A=Matrix([[ sqrt(n),0],[0,1]])
A=Matrix([[ cos(n),0],[0,1]])
A
```

```python
jones1= Matrix([0,1])
jones0= Matrix([1,0])
jones1
```

```python
prob=np.zeros(0)
A=A.subs(n,pi)
for i in range(1,200):
    B=Matrix([[cos(alpha),sin(alpha)],[-sin(alpha),cos(alpha)]])
    B=B.subs(N,i).evalf()
    Paola=(B*A)**(i-1)
    prob=np.append(prob,(jones1.T)*Paola*B*jones1)
    
```

```python
N=np.linspace(1,199,199)
plt.plot(N,np.square(prob),label='n=0.5')
plt.plot(N,np.square(prob1),label='n=0.2')
plt.plot(N,np.square(prob2),label='n=0.1')
plt.plot(N,np.square(prob3),label='n=0')
plt.plot(N,np.square(prob4),label='n=0.8')
plt.plot(N,np.square(prob5),label='n=1')


plt.xlabel('Numero de Beam Splitters')
plt.ylabel('Probabilidad de ser detectado en b')
plt.xlim(1,205)
plt.ylim(0,1.2)
plt.xticks(np.arange(1,202,25))
plt.legend()
```
